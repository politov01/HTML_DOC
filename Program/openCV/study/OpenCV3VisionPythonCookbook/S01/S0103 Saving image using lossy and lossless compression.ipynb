# Saving Images

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
import numpy as np

%matplotlib auto
%pylab inline


def print_image(header,name,np_arr,start_First=0, end_First=1, start_Second=0, end_Second=2):
    print(header)
    print('  Shape:', np_arr.shape, '  Data type:', np_arr.dtype)
    shapes = np_arr.shape #print(shapes)
    if shapes[0] < end_First:
        end_First = shapes[0]
    if shapes[1] < end_Second:
        end_Second = shapes[1]
    if len(shapes)==3:
        for i in range (start_First,end_First):
            print("[", end=" ")
            for k in range (start_Second,end_Second):
                print(np_arr[i,k], end=" ")
            if i == end_First-1:
                print(']  {}:'.format(name), np_arr.dtype)
            else:
                 print("]")
    if len(shapes)==2:
        for i in range (start_First,end_First):
            print("[", end=" ")
            for k in range (start_Second,end_Second):
                print(np_arr[i,k], end=" ")
            if i == end_First-1:
                print(']  {}:'.format(name), np_arr.dtype)
            else:
                print("]")

def plt_view_image(plt,list_images):
    #%pylab inline
    n = len(list_images)  #; print(n)
    plot_number = 1
    for name, img in list_images:
        plt.subplot(1,n,plot_number)
        plt.axis('off')
        plt.title(name)
        plt.imshow(img,cmap='gray' )
        plot_number = plot_number + 1
    plt.show()
#help("modules")   
#import sys             
#print('\n'.join(sys.path))
#pip list" 
```

# Reading the images from files

```python
#Let's import OpenCV module
import cv2

#Now we'll specify the path to our image
image_path = 'data/test.jpg'

#We'll use OpenCV's built-in function 'imread' to read the image
#Specify the image_path in the imread function
image = cv2.imread(image_path)

#If image is loaded successfully
#Go ahead and print the details
if image is not None:
    print('Image Loaded Successfully')
```

### Using the same image to flip

```python
#This function will flip the image on horizontal axis
horizontal_image = cv2.flip(image, 0)

#This function will flip the image on vertical axis
vertical_image = cv2.flip(image, 1)


#####################################################
print_image('original image','image',image,0,1,0,8)
print_image('horizontal_image','horizontal_image',horizontal_image,506,507,0,8)
print_image('vertical_image','vertical_image',vertical_image,0,1,631,639)

plt_view_image(plt,[["image",image[:,:,[2,1,0]] ], 
                    ["horizontal_image",horizontal_image[:,:,[2,1,0]] ],
                    ["vertical_image",vertical_image[:,:,[2,1,0]]   ]])
```

```python
# save image with lower compression - bigger file size but faster decoding
cv2.imwrite("horizontal_image_IMWRITE_PNG_COMPRESSION.png", horizontal_image, [cv2.IMWRITE_PNG_COMPRESSION, 0])
# save image with lower quality - smaller file size
cv2.imwrite("horizontal_image_IMWRITE_JPEG_QUALITY.jpg",    horizontal_image, [cv2.IMWRITE_JPEG_QUALITY, 0])
# save image with lower quality - smaller file size
cv2.imwrite("horizontal_image_IMWRITE_JPEG_QUALITY100.jpg", horizontal_image, [cv2.IMWRITE_JPEG_QUALITY, 100])
```

```python
# check that image saved and loaded again image is the same as original one
horizontal_image_IMWRITE_PNG_COMPRESSION = cv2.imread("horizontal_image_IMWRITE_PNG_COMPRESSION.png")
horizontal_image_IMWRITE_JPEG_QUALITY    = cv2.imread("horizontal_image_IMWRITE_JPEG_QUALITY.jpg")
horizontal_image_IMWRITE_JPEG_QUALITY100 = cv2.imread("horizontal_image_IMWRITE_JPEG_QUALITY100.jpg")


##########################################################
print_image('image','image',image,0,1,0,10)
print_image('horizontal_image','horizontal_image',horizontal_image,0,1,0,10)
print_image('horizontal_image_IMWRITE_PNG_COMPRESSION','horizontal_image_IMWRITE_PNG_COMPRESSION',horizontal_image_IMWRITE_PNG_COMPRESSION,0,1,0,10)
print_image('horizontal_image_IMWRITE_JPEG_QUALITY','horizontal_image_IMWRITE_JPEG_QUALITY',horizontal_image_IMWRITE_JPEG_QUALITY,0,1,0,10)
print_image('horizontal_image_IMWRITE_JPEG_QUALITY100','horizontal_image_IMWRITE_JPEG_QUALITY100',horizontal_image_IMWRITE_JPEG_QUALITY100,0,1,0,10)

plt.figure(figsize=(18,18))
plt_view_image(plt,[["image",image[:,:,[2,1,0]] ], 
                    ["horizontal_image",horizontal_image[:,:,[2,1,0]] ],
                    ["horizontal_image_IMWRITE_PNG_COMPRESSION",horizontal_image_IMWRITE_PNG_COMPRESSION[:,:,[2,1,0]]  ],
                   ])
plt.figure(figsize=(18,18))
plt_view_image(plt,[["image",image[:,:,[2,1,0]] ], 
                    ["horizontal_image_IMWRITE_JPEG_QUALITY",horizontal_image_IMWRITE_JPEG_QUALITY[:,:,[2,1,0]]   ],
                    ["horizontal_image_IMWRITE_JPEG_QUALITY100",horizontal_image_IMWRITE_JPEG_QUALITY100[:,:,[2,1,0]]   ]
                   ])


cv2.imshow('image, horizontal_image',
           np.hstack([image, horizontal_image,horizontal_image_IMWRITE_PNG_COMPRESSION,horizontal_image_IMWRITE_JPEG_QUALITY]))
cv2.imshow('horizontal_image, horizontal_image_IMWRITE_PNG_COMPRESSION',
           np.hstack([horizontal_image,horizontal_image_IMWRITE_PNG_COMPRESSION]))
cv2.imshow('horizontal_image, horizontal_image_IMWRITE_JPEG_QUALITY',
            np.hstack([ horizontal_image,horizontal_image_IMWRITE_JPEG_QUALITY]))
cv2.imshow('horizontal_image, horizontal_image_IMWRITE_JPEG_QUALITY100',
            np.hstack([ horizontal_image,horizontal_image_IMWRITE_JPEG_QUALITY100]))
cv2.waitKey()
cv2.destroyAllWindows()
```
