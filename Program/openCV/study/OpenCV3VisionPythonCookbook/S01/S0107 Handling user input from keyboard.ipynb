# Handling User Input from Keyboard

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib auto
%pylab inline


def print_image(header,name,np_arr,start_First=0, end_First=1, start_Second=0, end_Second=2):
    print(header)
    print('  Shape:', np_arr.shape, '  Data type:', np_arr.dtype)
    shapes = np_arr.shape #print(shapes)
    if shapes[0] < end_First:
        end_First = shapes[0]
    if shapes[1] < end_Second:
        end_Second = shapes[1]
    if len(shapes)==3:
        for i in range (start_First,end_First):
            print("[", end=" ")
            for k in range (start_Second,end_Second):
                print(np_arr[i,k], end=" ")
            if i == end_First-1:
                print(']  {}:'.format(name), np_arr.dtype)
            else:
                 print("]")
    if len(shapes)==2:
        for i in range (start_First,end_First):
            print("[", end=" ")
            for k in range (start_Second,end_Second):
                print(np_arr[i,k], end=" ")
            if i == end_First-1:
                print(']  {}:'.format(name), np_arr.dtype)
            else:
                print("]")

def plt_view_image(plt,list_images):
    #%pylab inline
    n = len(list_images)  #; print(n)
    plot_number = 1
    for name, img in list_images:
        plt.subplot(1,n,plot_number)
        plt.axis('off')
        plt.title(name)
        plt.imshow(img,cmap='gray' )
        plot_number = plot_number + 1
    plt.show()
#help("modules")   
import sys             
print('\n'.join(sys.path))
#pip list" 
```

# Reading the images from files

```python
#Let's import OpenCV module
import cv2

#Now we'll specify the path to our image
image_path = 'data/test.jpg'

#We'll use OpenCV's built-in function 'imread' to read the image
#Specify the image_path in the imread function
image = cv2.imread(image_path)

#If image is loaded successfully
#Go ahead and print the details
if image is not None:
    print('Image Loaded Successfully')
```

```python
# Let's first create a namedWindow
# and display the image that we just read

cv2.namedWindow('image')
cv2.imshow('image', image)

# We'll then wait for the user to press a key
key = cv2.waitKey()

# We'll display the unicode code of the key pressed
print(key)

# and now we'll flip the image depending upon user input
# If user presses 'h' we'll horizontally flip the image
# and otherwise vertically

# The ord() function returns the unicode code 
# representing the character 

if key == ord('h'):
    #This function will flip the image on horizontal axis
    image_h = cv2.flip(image, 0)
    cv2.imshow('image_h', image_h)
    key = cv2.waitKey()
    cv2.destroyAllWindows()

if key == ord('v'):
    #This function will flip the image on vertical axis
    image_v = cv2.flip(image, 1)
    cv2.imshow('image_v', image_v)
    key = cv2.waitKey()
    cv2.destroyAllWindows()

# Let's say we want to save this image,
# We'll just press 's' key and the snippet 
# will store the image in the current directory
if key == ord('s'):
    cv2.imwrite('image_we_saved.jpg', image)
    print('Saved!')
else:
    print('Image not Saved!')

plt.figure(figsize=(18,18))
plt_view_image(plt,[["image",image[:,:,[2,1,0]] ], 
                    ["image_h",image_h[:,:,[2,1,0]] ],
                    ["image_v",image_v[:,:,[2,1,0]]  ]
                   ])

cv2.waitKey()
cv2.destroyAllWindows()
```
